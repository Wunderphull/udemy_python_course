```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
```

```python
# We are going to pass a dicitonary in this DataFrame:

dframe1 = DataFrame({'key':['X','Z','Y','Z','X','X'],'data_set_1': np.arange(6)})

dframe1
```

```python
dframe2 = DataFrame({'key':['Q','Y','Z'],'data_set_2':[1,2,3]})

dframe2
```

```python
# Now we may use the 'merge' method. This is a many-to-one method:
# It will automatically chose overlapping columns to merge on.

pd.merge(dframe1,dframe2)
```

```python
# The only similarities between the dataframes are Z and Y. There were no overlapping X.
```

```python
# Match on a specific column:

pd.merge(dframe1,dframe2,on='key')
```

```python
# We can choose which DataFrame's keys to use, this will choose left (dframe1):

pd.merge(dframe1,dframe2,on='key',how='left')
```

```python
# We can choose which DataFrame's keys to use, this will choose left (dframe2):


pd.merge(dframe1,dframe2,on='key',how='right')
```

```python
#Choosing the "outer" method selects the union of both keys

pd.merge(dframe1,dframe2,on='key',how='outer')
```

```python
# Now we will conduct a many-to-many merge:
```

```python
dframe3 = DataFrame({'key': ['X', 'X', 'X', 'Y', 'Z', 'Z'],
                 'data_set_3': range(6)})
dframe4 = DataFrame({'key': ['Y', 'Y', 'X', 'X', 'Z'],
                 'data_set_4': range(5)})

dframe3
```

```python
dframe4
```

```python
pd.merge(dframe3, dframe4)

# So what happened? A many to many merge results in the product of the rows.
# Because there were 3 'X's in dframe3 and 2 'X's in dframe4 there ended up being a total of 6 'X' rows in the result
# (2*3=6)! 
# Note how dframe3 repeats its 0,1,2 values for 'X' and dframe4 repeats its '2,3' pairs throughout the key set.
```

```python
# Merge with multiple keys:

# Dframe on left
df_left = DataFrame({'key1': ['SF', 'SF', 'LA'],
                  'key2': ['one', 'two', 'one'],
                  'left_data': [10,20,30]})

#Dframe on right
df_right = DataFrame({'key1': ['SF', 'SF', 'LA', 'LA'],
                   'key2': ['one', 'one', 'one', 'two'],
                   'right_data': [40,50,60,70]})
```

```python
df_left
```

```python
df_right
```

```python
#Merge

pd.merge(df_left, df_right, on=['key1', 'key2'], how='outer')
```

```python
#Note that the left and right DataFrames have overlapping key names (key1 and key2).
# pandas automatically adds suffixes to them.

pd.merge(df_left,df_right,on='key1')
```

```python
# We can also specify what the suffix becomes:

pd.merge(df_left,df_right, on='key1',suffixes=('_lefty','_righty'))
```
