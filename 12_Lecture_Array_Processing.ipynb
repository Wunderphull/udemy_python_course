```python
# Numpy Where:

A = np.array([1,2,3,4])

B = np.array([100,200,300,400])
```

```python
condition = np.array([True,True,False,False])
```

```python
'''
Outputs the value of A if the condition is met, otherwise it outputs the value of B.
This means that, if the value of A is '1' and the condition is set to TRUE, then the value will be '1'.
If the condition is set to FALSE, then it will output the value of B.

'''

answer = [(A_val if cond else B_val) for A_val,B_val,cond in zip(A,B,condition)]
```

```python
answer
```

```python
# This is a shortform of the above function.
answer2 = np.where(condition,A,B)
```

```python
answer2
```

```python
condition = np.array([False,True,True,False])
```

```python
answer3 = np.where(condition,A,B)

answer3
```

```python
answer4 = np.where(condition,B,A)

answer4
```
